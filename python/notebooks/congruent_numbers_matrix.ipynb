```python
import numpy as np
import csv
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sympy import sieve
import itertools
import pandas as pd
```

```python
matrix_size = 200
matrix = np.zeros((matrix_size, matrix_size))
with open('C:/Users/esultano/git/elliptic_curves/data/elliptic_curves.txt') as f:
    for line in f.readlines():
        line = line.strip()
        p, q, cases = eval(line)
        idx_p = sieve.search(p)[0]-2
        idx_q = sieve.search(q)[0]-2
        if idx_p < matrix_size and idx_q < matrix_size:
            cases_set = set()
            for case in cases:
                cases_set.add(case[0])
            cases_list = list(cases_set)
            val = sum(i*i for i in cases_list)
            matrix[idx_p,idx_q] = val
            matrix[idx_q,idx_p] = val

plt.figure(figsize=(8, 8))
plt.matshow(matrix, fignum=1)
plt.show()
```
